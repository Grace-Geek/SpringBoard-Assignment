```python
# Summary statistics for numerical features
numerical_features = df.select_dtypes(include=[np.number])
summary_stats = numerical_features.describe().T

display(summary_stats)
```

```python
# Summary statistics for categorical features
categorical_features = df.select_dtypes(include=[object])
cat_summary_stats = categorical_features.describe().T

display(cat_summary_stats)
```

```python
# Null values in the dataset
null_values = df.isnull().sum()
null_values_nonzero = null_values[null_values > 0].sort_values(ascending=False)
# Percentage of missing values for each feature
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage_nonzero = missing_percentage[missing_percentage>0].sort_values(ascending=False)
display(null_values_nonzero, missing_percentage_nonzero)
```

```python
rows_with_missing_values = df[df.isnull().any(axis=1)]
display(rows_with_missing_values)
```
